# 문제 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_total = pd.read_excel(r'starbucks_result_06.xlsx')
s=raw_total['hashtag']
s=s.str.replace(" ","")
s=s.astype(str)
```

```python
tags_total = []

for tags in s:
    tags_list = tags[2:-2].split("','")
    for tag in tags_list:
        tags_total.append(tag)
```

```python
tags_total
```

```python
#Bi-gram
list(zip(tags_total))
```

```python
#tri-gram
list(zip(tags_total, tags_total[1:], tags_total[3:])) 
```

```python
#4-grams
list(zip(tags_total, tags_total[1:], tags_total[3:], tags_total[4:])) 
```

```python
#5-grams
list(zip(tags_total, tags_total[1:], tags_total[3:], tags_total[4:], tags_total[5:]))
```

```python
# !pip install gensim newspaper3k
# !pip install gensim==3.8.1
from gensim.summarization.summarizer import summarize
from newspaper import Article

url = "https://finance.yahoo.com/news/apple-ios-15-debut-174727899.html"
news = Article(url, language='en')
news.download()
news.parse()
print(news.text)
```

```python
# 기본 요약
print(summarize(news.text))
```

```python
# 출력단어 500개로 제한
print(summarize(news.text, word_count=500))
```

```python
# 문장비율 설정 요약
print(summarize(news.text, ratio=0.1))
```
